```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
yes = 0 #counter for number of times we land on 25
target = 25 #step 25
runs = 10000000 #repeat simulation this many times
expected = (2 / 3) * (1 - np.power((-0.5), target)) #probability of landing on target step using difference equation formula

for i in range(runs):
    current = 1 #start at step 1
    done = False
    while done == False: #while we have not arrived at/surpassed step 25
        coin = random.choice((1, 2)) #flip coin
        current += coin #move forward
        if current == target: #if on step 25
            yes += 1
            done = True
        if current == target + 1: #if we skipped 25 and are now on 26
            done = True

actual = yes / runs #estimated probability
print("Total number of simulations: " + str(runs))
print("Number of times we landed on " + str(target) + ": " + str(yes))
print("Expected probability: " + str(expected))
print("Actual probability: " + str(actual))
print("Inaacuracy: " + str((np.abs(expected - actual) / expected) * 100) + "%")
```

```python
x = np.arange(1, 15)
y = (2 / 3) * (1 - np.power(-0.5, x))
plt.plot(x, y)
plt.title('A graph to show the probability that we land on any given step during the simulation')
plt.xlabel('Step (n)')
plt.ylabel('p(n)')
plt.show()
```
